```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

beta = 2.75
f = 0.05
delta = 10
c = 0.5
T=50;
x0 = [1, 0]
def F(t, x):
    S = x[0]
    K = x[1]
    F = np.empty(2)
    F[0] = (1-S) * S - S * K / (c + beta * K)
    F[1] = delta * ( c * S / 2/ (c + beta* S * K)**2 - f)
    return F

t_eval = np.arange(0, T, 0.01)
sol = solve_ivp(F, [0, T], [1, 0], t_eval = t_eval)

plt.figure(figsize =(12,8))
plt.plot(sol.t, sol.y.T[:,0]) # t is column vector, sol.y is row vector so transpose is necessary.
plt.xlabel('time')
plt.ylabel('S: Fish population')
plt.show()

plt.figure(figsize =(12,8))
plt.plot(sol.t, sol.y.T[:,1]) # t is column vector, sol.y is row vector so transpose is necessary.
plt.xlabel('time')
plt.ylabel('K: number of firms')
plt.show()


plt.figure(figsize =(12,8))
plt.plot(sol.y.T[:,0], sol.y.T[:,1]) # t is column vector, sol.y is row vector so transpose is necessary.
plt.xlabel('S')
plt.ylabel('K')
plt.show()
```
